###  **Subplots**
---

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a theme
sns.set_theme(style="whitegrid")

# Load a dataset
dm = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "WS2", "WS1", "IF"]
sns.scatterplot(x="carat", y="price", hue="clarity", size="depth",
                palette="ch:r=-2,d=.3_r", hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0, data=dm, ax=ax)

plt.show()
```

###  **Boxplot and Stripplot** 
#### **(Merge)**
---

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a theme
sns.set_theme(style="ticks")

#  Initialize the  figure with a Logarithmic x axis
f, ax = plt.subplots(figsize=(7, 6))

# Load a dataset
dm = sns.load_dataset("diamonds")

# Plot the orbital period with horizontal boxes
sns.boxplot(x="distance", y="method", data="dm",
            whis=[0, 100], width=.6, palette="vlag")

#  Add in points to show each observation
sns.stripplot(x="distance", y="method", data="dm",
              size=4, color=".1", linewidth=0)
# Tweak the visual presentation 
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)

plt.show()
```
