### **_1st step import libraries_**

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

### **_2nd step set a theme_**
---

```python
sns.set_theme(style="ticks", color_codes=True)
```

### **_3rd import dataset you can also import your own data_**
---

```python
kashti= sns.load_dataset("titanic")
kashti
# kashti.info
# kashti.shape
# kashti.describe
```

### **_4th step plot basic graph(count plot)_**
---

```python
# countplot ki left side mein already cout aa jata hai we will not put any y exis here .
k=sns.countplot(x="sex", data=kashti)
plt.show()
```

### **_5th step plot graph with two variables (count plot)with title_**
---

```python
k=sns.countplot(x="sex", data=kashti,hue="class")
plt.title("--Kashti--")
plt.show()
```
