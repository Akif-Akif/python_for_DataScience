```python
# Steps involved in Data visualization
# step_1 import libraries
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# # step_2 set a theme
sns.set_theme(style="ticks",color_codes=True)

# step_3 import data set you can also import your own dat
df= sns.load_dataset("titanic")
df
```

```python
#step_4 plot basic graph with 1 variable (count)
sns.countplot(x= "sex", data=df)
plt.show()
```

```python
# Step_5 plot basic graph with 2 variables |(count plot) with titles
df = sns.countplot(x="sex",  hue="class", data=df)
df.set_title("Plot for basic Counting")
plt.show()
```
