# _**Pandas_history and Introduction**_

Pandas is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis"
 and was created by Wes McKinney in 2008.

- Development of Pandas started from 2008
- Pandas become open source in 2009
- First edition of Python for data analysis is published 2012
- Pandas becomes a NUMFOCUS sponsored project (NUMFOCUS aik company thi jis ne isy sponsored kya tha)
- First in-person core developer sprint 



### **Why Pandas ?**
Pandas allows us to analyze big data and make conclusions based on statistical theories.

Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.
